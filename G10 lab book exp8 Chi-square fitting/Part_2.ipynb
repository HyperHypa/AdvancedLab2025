```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2
```

```python

np.random.seed(0)          # reproducible results
n = 500            # number of samples
mu, sigma = 0, 5            # population parameters

data = np.random.normal(loc=mu, scale=sigma, size=n)
chisq = ((data - mu) / sigma) ** 2
```

```python
# Plot histogram + theoretical pdf (df=1)
plt.figure() # create a new figure
plt.hist(chisq, bins=30, density=True, alpha=0.6) # histogram
x = np.linspace(0, np.percentile(chisq, 100), 300) # x-axis values
plt.plot(x, chi2.pdf(x, df=1), linewidth=2) # theoretical pdf
plt.title("Chi‑square values (df = 1)") # title of the plot 
plt.xlabel("χ² value") 
plt.ylabel("Probability density")
plt.grid(True)
plt.show()
```

```python
DoF = 1000
for k in range(DoF,DoF+1):
    # k = 10                       # number of independent arrays (change 2–10 to explore)\
   
    arrays = np.random.normal(loc=mu, scale=sigma, size=(k, n))
    chisq_arrays = ((arrays - mu) / sigma) ** 2
    sum_array = np.sum(chisq_arrays, axis=0)    # χ²(df=k) for each of the 500 positions

    # Plot histogram + theoretical pdf (df=k)
    plt.figure() # create a new figure
    plt.hist(sum_array, bins=30, density=True, alpha=0.6) # histogram
    x = np.linspace(0, np.percentile(sum_array, 100), 300) # x-axis values
    plt.plot(x, chi2.pdf(x, df=k), linewidth=2) # theoretical pdf
    plt.title(f"Sum of {k} Independent Squared Variables (DoF = {k})") # title of the plot
    plt.xlabel("χ² value")
    plt.ylabel("Probability density")
    #plt.xlim(800, 1200)
    plt.grid(True)
    plt.show()
```
